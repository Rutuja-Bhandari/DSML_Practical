```python
import pandas as pd
df=pd.read_csv("D:\Datasets\Datasets\Lipstick.csv")
df=df.drop(columns=['Id'])
df
# dataset=pd.read_csv("D:\Datasets\Datasets\Lipstick.csv")
# dataset=dataset.drop(columns=['Id'])
# dataset
```

```python
#0=<21, 1=21-35, 2=>35
def age(value):
    if(value=="<21"):
        return 0
    elif(value=="21-35"):
        return 1
    else:
        return 2
    
df['Age']=df['Age'].apply(age)
```

```python
#0=low, 1=medium, 2=high
def income(value):
    if(value=="High"):
        return 2
    elif(value=="Medium"):
        return 1
    else:
        return 0
    
df['Income']=df['Income'].apply(income)
```

```python
#0=male, 1=female
def Gender(value):
    if(value=="Male"):
        return 0
    elif(value=="Female"):
        return 1
    
df['Gender']=df['Gender'].apply(Gender)
```

```python
#0=single, 1=married
def Ms(value):
    if(value=="Single"):
        return 0
    elif(value=="Married"):
        return 1
    
df['Ms']=df['Ms'].apply(Ms)
```

```python
#0=no, 1=yes
def Buys(value):
    if(value=="Yes"):
        return 1
    elif(value=="No"):
        return 0
    
df['Buys']=df['Buys'].apply(Buys)
```
