```python
from sklearn.cluster import KMeans
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.DataFrame({'X':[0.1,0.15,0.08,0.16,0.2,0.25,0.24,0.3],'Y':[0.6,0.71,0.9,0.85,0.3,0.5,0.1,0.2]})
f1=df['X'].values
f2=df['Y'].values
X=np.array(list(zip(f1,f2)))
print(X)
```

```python
C_x=np.array([0.1,0.6])
C_y=np.array([0.3,0.2])
centroid=C_x,C_y
print(np.array(list((C_x,C_y))))
```

```python
plt.scatter(f1,f2)
```

```python
plt.scatter(C_x[0],C_x[1],color='red')
plt.scatter(C_y[0],C_y[1],color='green')
```

```python
plt.scatter(f1,f2)
plt.scatter(C_x[0],C_x[1],marker='*',s=200,color='red')
plt.scatter(C_y[0],C_y[1],marker='*',s=200,color='green')
```

```python
model=KMeans(n_clusters=2,random_state=0)
model.fit(X)
labels=model.labels_
for point,labels in zip(X,labels):
    print("points: ",point," ","label: ",labels)
```

```python
new_cluster=model.cluster_centers_
print("M1= ",new_cluster[0]," M2= ",new_cluster[1])
```

```python
plt.scatter(new_cluster[0][0],new_cluster[0][1])
plt.scatter(new_cluster[1][0],new_cluster[1][1])
plt.scatter(C_x[0],C_x[1],s=200,marker='*')
plt.scatter(C_y[0],C_y[1],s=200,marker='*')
```
