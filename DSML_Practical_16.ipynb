16. Use the inbuilt dataset 'titanic'. The dataset contains 891 rows and
contains information about the passengers who boarded the unfortunate
Titanic ship. Write a code to check how the price of the ticket (column
name: 'fare') for each passenger is distributed by plotting a histogram.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
titanic_data=sns.load_dataset('titanic')
titanic_data
```

```python
sns.histplot(x='fare',data=titanic_data,color='blue',kde='true')
plt.show()
```

```python
sns.scatterplot(x='fare',data=titanic_data,y='sex')
plt.show()
```

```python
sns.countplot(x='alone',hue='fare',data=titanic_data)
```

```python
# 4. Scatter plot for fare and age
plt.figure(figsize=(10, 6))
sns.scatterplot(x='age', y='fare', hue='survived', data=titanic_data, palette='coolwarm')
plt.title('Scatter plot of Fare vs. Age with Survival')
plt.show()
```

```python
sns.histplot(data=titanic_data['fare'],color='navy',bins=5,alpha=1)

plt.show()
```
