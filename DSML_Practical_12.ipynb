```python
# Function to identify and plot outliers
def plot_outliers(column_data, column_name):
    plt.figure(figsize=(8, 6))
    
    # Calculate IQR
    Q1 = np.percentile(column_data, 25)
    Q3 = np.percentile(column_data, 75)
    IQR = Q3 - Q1
    
    # Define the lower and upper bounds for outliers
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Identify outliers
    outliers = column_data[(column_data < lower_bound) | (column_data > upper_bound)]
    
    # Plot the boxplot
    plt.boxplot(column_data, vert=False)
    
    # Highlight the outliers
    plt.scatter(outliers, [1] * len(outliers), color='red', label='Outliers')
    
    plt.title(f'Outliers in {column_name}')
    plt.xlabel(column_name)
    plt.yticks([])
    plt.legend()
    plt.show()

```

```python
plot_outliers(df['sepal_length'], 'sepal_length')
```

```python
plot_outliers(df['sepal_width'], 'sepal_width')
```

```python
plot_outliers(df['petal_length'], 'petal_length')
```

```python
plot_outliers(df['petal_width'], 'petal_width')
```
