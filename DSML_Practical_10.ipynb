10. Write a program to do the following: You have given a collection of 8
points. P1=[2, 10] P2=[2, 5] P3=[8, 4] P4=[5, 8] P5=[7,5] P6=[6, 4] P7=[1, 2]
P8=[4, 9]. Perform the k-mean clustering with initial centroids as m1=P1
=Cluster#1=C1 and m2=P4=cluster#2=C2, m3=P7 =Cluster#3=C3. Answer
the following 1] Which cluster does P6 belong to? 2] What is the
population of a cluster around m3? 3] What is the updated value of m1,
m2, m3?


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
df=pd.DataFrame({'X':[2,2,8,5,7,6,1,4],'Y':[10,5,4,8,5,4,2,9]})
f1=df['X'].values
f2=df['Y'].values
X=np.array(list(zip(f1,f2)))
print(X)
```

```python
C_x=np.array([2,10])
C_y=np.array([5,8])
C_z=np.array([1,2])
print(C_x,C_y,C_z)
```

```python
plt.scatter(f1,f2)
```

```python
plt.scatter(C_x[0],C_x[1])
plt.scatter(C_y[0],C_y[1])
plt.scatter(C_z[0],C_z[1])
```

```python
model=KMeans(n_clusters=3,random_state=0)
model.fit(X)
labels=model.labels_
for point,label in zip(X,labels):
    print("Point:",point," ","label:",label)
```

```python
new_centroid=model.cluster_centers_
print(new_centroid)
```

```python
plt.scatter(new_centroid[0][0],new_centroid[0][1],marker='*',color='blue',s=200)
plt.scatter(new_centroid[1][0],new_centroid[1][1],marker='*',color='blue',s=200)
plt.scatter(new_centroid[2][0],new_centroid[2][1],marker='*',color='blue',s=200)

plt.scatter(C_x[0],C_x[1],marker='*',color='purple',s=200)
plt.scatter(C_y[0],C_y[1],marker='*',color='purple',s=200)
plt.scatter(C_z[0],C_z[1],marker='*',color='purple',s=200)

plt.scatter(f1,f2)
```
