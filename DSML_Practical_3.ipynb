```python
def convert_to_integer(value):
    try:
        return int(value)
    except ValueError:
       return 0 
df['GrossSquareMeters']=df['GrossSquareMeters'].str.replace('m2','')
df['GrossSquareMeters']=df['GrossSquareMeters'].apply(convert_to_integer)
```

```python
def convert_to_integer(value):
    try:
        return int(value)
    except ValueError:
       return 0 
df['price']=df['price'].str.replace('TL','')
df['price']=df['price'].str.replace(',','')
df['price']=df['price'].apply(convert_to_integer)
```

```python
def convert_to_integer(value):
    try:
        return int(value)
    except ValueError:
       return 0 
df['NetSquareMeters']=df['NetSquareMeters'].str.replace('m2','')
df['NetSquareMeters']=df['NetSquareMeters'].apply(convert_to_integer)
```

```python
df
```

```python
for column in df.columns:
    if(df[column].convert_dtypes().dtypes!='string'):
        print(column," ",df[column].mean())
```

```python
for column in df.columns:
    if(df[column].convert_dtypes().dtypes!='string'):
        print(column," ",df[column].median())
```

```python
for column in df.columns:
    if(df[column].convert_dtypes().dtypes!='string'):
        print(column," ",df[column].std())
```

```python
for column in df.columns:
    if(df[column].convert_dtypes().dtypes!='string'):
        print(column," ",df[column].var())
```

```python
for column in df.columns:
    if(df[column].convert_dtypes().dtypes!='string'):
        percentile=df[column].quantile([0.25,0.5,0.75])
        print(column)
        print("25= ",percentile[0.25])
        print("50= ",percentile[0.50])
        print("75= ",percentile[0.75])
        print()
```
