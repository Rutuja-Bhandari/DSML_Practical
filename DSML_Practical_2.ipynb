```python

for column in df.columns:
    if(df[column].convert_dtypes().dtypes!='string' and df[column].convert_dtypes().dtypes!='boolean'):
        percentile=df[[column]].quantile([0.25,0.5,0.75])
    
    print("25% percentile ",percentile.loc[0.25])
    print("50% percentile ",percentile.loc[0.50])
    print("75% percentile ",percentile.loc[0.75])
```
